```python
# 'Booking Date' and 'Arrival Date' are in datetime format
df['Booking Day'] = df['Booking Date'].dt.day_of_week
df['Arrival Day'] = df['Arrival Date'].dt.day_of_week
df['Checkout Day'] = df['Status Update'].dt.day_of_week

# Map week numbers to week names
df['Booking Day'] = df['Booking Day'].apply(lambda x: calendar.day_name[x])
df['Arrival Day'] = df['Arrival Day'].apply(lambda x: calendar.day_name[x])
df['Checkout Day'] = df['Checkout Day'].apply(lambda x: calendar.day_name[x])

week_order = list(calendar.day_name)
df['Booking Day'] = pd.Categorical(df['Booking Day'], categories=week_order, ordered=True)
df['Arrival Day'] = pd.Categorical(df['Arrival Day'], categories=week_order, ordered=True)
df['CheckOut Day'] = pd.Categorical(df['Checkout Day'], categories=week_order, ordered=True)

df_filtered = df[df['Status'] == 'Check-Out']

# Explore the Comparison of Booking and Arrival Day
plt.figure(figsize=(10, 5))
sns.lineplot( df_filtered['Booking Day'].value_counts(sort=False),  label='Booking')
sns.lineplot(df_filtered['Arrival Day'].value_counts(sort=False),  label='Arrival')
sns.lineplot( df_filtered['CheckOut Day'].value_counts(sort=False),  label='Check Out')

plt.title('Comparison of Booking and Arrival Day')
plt.xlabel('')
plt.ylabel('')
plt.legend()
plt.show()

```

The booking, arrival, and check-out patterns suggest that the majority of activities occur from Monday to Friday, with a peak in bookings on Wednesdays. Fridays seem to be popular for both bookings and arrivals. Additionally, check-outs are more frequent on Sundays, indicating that many guests prefer to conclude their stays over the weekend. This information can help optimize staff scheduling and resource allocation during different days of the week.

```python
# 'Booking Date' and 'Arrival Date' are in datetime format
df['Booking Month'] = df['Booking Date'].dt.month
df['Arrival Month'] = df['Arrival Date'].dt.month

# Map month numbers to month names
df['Booking Month'] = df['Booking Month'].apply(lambda x: calendar.month_name[x])
df['Arrival Month'] = df['Arrival Month'].apply(lambda x: calendar.month_name[x])

# Filter data for years other than 2017
df_filtered = df[df['Booking Date'].dt.year < 2017]

# Set the 'Booking Month' column as an ordered categorical variable
month_order = [calendar.month_name[i] for i in range(1, 13)]
df_filtered['Booking Month'] = pd.Categorical(df_filtered['Booking Month'], categories=month_order, ordered=True)
df_filtered['Arrival Month'] = pd.Categorical(df_filtered['Arrival Month'], categories=month_order, ordered=True)

plt.figure(figsize=(12, 8))

# Create a grouped bar chart for booking and arrival months
bar_width = 0.35
bar_positions_booking = np.arange(len(month_order))
bar_positions_arrival = bar_positions_booking + bar_width

plt.bar(bar_positions_booking, df_filtered['Booking Month'].value_counts(sort=False), width=bar_width, label='Booking')
plt.bar(bar_positions_arrival, df_filtered['Arrival Month'].value_counts(sort=False), width=bar_width, label='Arrival')

plt.title('Comparison of Booking and Arrival Months Over Time (Excluding 2017)')
plt.xlabel('Month')
plt.ylabel('Count')
plt.xticks(bar_positions_booking + bar_width / 2, month_order)
plt.legend()
plt.show()
```

January and February emerge as peak booking months, with January leading in counts. There is a steady increase in bookings from April to August, peaking in July. Notably, October stands out with the highest booking and arrival counts, suggesting a significant peak in travel during this month. The decline in bookings in November and December aligns with the holiday season. It's important to note that the data for 2017 is excluded due to incomplete records

```python
# Plot the distribution of Country by Revenue
top_countries = df.groupby('Country')['Revenue'].sum().reset_index().nlargest(10, 'Revenue')

plt.figure(figsize=(12, 6))
sns.barplot(x='Revenue', y='Country', data=top_countries)
plt.title('Top 10 Countries by Revenue ')
plt.xlabel('Revenue')
plt.ylabel('')
plt.show()
```

Portugal stands out as the highest revenue generator, contributing significantly more than other countries. To optimize revenue, strategies should focus on enhancing offerings and customer engagement in Portugal.

```python
countries_loss = df.groupby('Country')['Revenue Loss'].sum().reset_index().nsmallest(10, 'Revenue Loss')
plt.figure(figsize=(12, 6))
ax = sns.barplot(x='Revenue Loss', y='Country', data=countries_loss)

plt.title('Top 10 Countries by Revenue Loss (Stacked by Status)')
plt.xlabel('Revenue Loss')
plt.ylabel('')
plt.show()

```

Portugal not only leads in revenue generation but also faces the highest revenue loss among the top 10 countries. Identifying and addressing factors contributing to cancellations in Portugal could significantly impact profitability. Moreover, attention should be directed towards mitigating losses in the United Kingdom, Spain, France, and Germany to enhance overall revenue performance.

```python
df['Profit'] =df['Revenue'] + df['Revenue Loss']
countries_profit = df.groupby('Country')['Profit'].sum().reset_index().nlargest(10, 'Profit')
plt.figure(figsize=(12, 6))
ax = sns.barplot(x='Profit', y='Country', data=countries_profit, dodge=False)

plt.title('Top 10 Countries by Profit')
plt.xlabel('Revenue Loss')
plt.ylabel('')
plt.show()

```

Portugal maintains its position as the most profitable country, contributing significantly to the overall profit. The United Kingdom, France, Germany, and Spain also demonstrate substantial profitability. Strategically reinforcing marketing efforts and customer engagement in these countries could further enhance overall profits. 

```python
# Count the bookings for each combination of 'Lead Time' and 'Status'
booking_counts = df.groupby(['Lead Time', 'Status']).size().reset_index(name='Booking Count')
booking_counts = booking_counts[booking_counts['Status'] == 'Check-Out']
plt.figure(figsize=(12, 6))
sns.lineplot(x='Lead Time', y='Booking Count', data=booking_counts)
plt.title('How Lead Time Affects the Status of Customers (Check-Out)')
plt.xlabel('Lead Time (days)')
plt.ylabel('Count')
plt.show()
```

 As lead time increases, the  count shows a declining trend. This suggests that customers with shorter lead times are more likely to complete their bookings and proceed to check-out. 

## Recommendations

1. **December Arrival Boost:**
    December experiences high bookings but doesn't translate into the highest arrivals. Strategize promotions or incentives to enhance arrivals during the holiday season.

2. **Optimizing for October Peaks:**
   Leverage the popularity of October by tailoring promotions or events to align with the peak travel period, maximizing both bookings and arrivals.

3. **Flexible Deposit Options:**
    Customers with non-refundable deposits show higher cancellation rates. Consider offering flexible deposit options to address this trend.

4. **Last-Minute Deals:**
    Capture spontaneous bookings by implementing targeted last-minute promotions, attracting customers with shorter lead times.
5. **Evaluate Travel Agent Impact:**
   While utilizing travel agents proves effective, investigate and address factors leading to cancellations post travel agent bookings.

6. **Tailoring Services for Contract Customers:**
   "Contract" customers tend to have notably longer stays. Understand their unique needs and tailor services to enhance their experience.

7. **Strategic Resource Allocation:**
Optimize staffing levels for weekdays with a focus on Wednesdays, the peak day for bookings. Additionally, allocate resources for increased activity during Sunday check-outs

8.   **Weekend Check-Out Emphasis:**
   Given the preference for Sunday check-outs, focus on enhancing services and promotions during the weekends to attract guests looking for a relaxing conclusion to their stay.

9. **Explore Strategies for Portugal:**
Portugal appears to be a significant contributor to both bookings and cancellations. To improve profitability, explore strategies to reduce cancellations in Portugal, such as personalized promotions or loyalty programs

10. **Diversify Revenue Sources:**
Exploring opportunities for revenue growth in other countries by considering increasing advertising efforts in countries beyond Portugal, the UK, France, Spain, and Germany, could contribute to diversification and risk mitigation.

These recommendations aim to optimize operations, enhance customer experience, and refine marketing strategies based on data-driven insights. Consider implementing these strategies gradually and monitor their impact on key performance indicators.
